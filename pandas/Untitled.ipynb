```python
survive_female = titanic_df[
    (titanic_df["Sex"] == "female") & (titanic_df["Survived"] == 1)
]
```

```python
len(survive_male)/len(titanic_df)
```

```python
passanger_class_2nd_survive = titanic_df[
    (titanic_df["Pclass"] == 2) & (titanic_df["Survived"] == 1)
    ]
passanger_class_2nd = titanic_df[
    titanic_df["Pclass"] == 2 ]
len(passanger_class_2nd_survive)/len(passanger_class_2nd)
```
